```python
"""
Created on Jan 26 2022

Author: Sejzi Remorque
"""

"""
    DocString:
    
    A) Introduction:
    This game is based on the books and TV series 'The Witcher' by Andrzej Sapkowski(1986).
    This particular story is retrieved from one of the short story collections called 
    "The Witcher". This game is story driven to help those who did not read or watched
    the series but will still be able to enjoy and understand what's going on.
    It consists of 4 stages and has defined functions with loops for replayability
    and sequence matching to engage players with the story.
    
    
    
    Round 1: Decide where to gather information.
    Round 2: Decide your approach to the agressive thugs
    Round 3: Arrange your destination in the right order
    Round 4: Fight the monster until it is weakened
    
    B) Known Bugs and/or Errors:
    None.
    
"""

from sys import exit
import random

def start():
    #######Just a little background story about your character's nature
    print("""
You are journeying as Geralt of Rivia, known as the White Wolf, and a legendary amongst witchers. These Witchers are
offspring of witchcraft and sorcery solely created as mankind's greatest defenders against beast and monsters. As the 
golden age of witchers has come to pass, monsters grew increasingly rare. Due to Churches' and ruler's propagandas,
witchers became the monsters in populations eyes. Being a mutant means you are despised and hated by the common people. 
But witchers live by the code and it is your duty to protect the innocents and help those in need.
""")

    input(prompt ="<Press enter to continue>\n") 
    
    ######Start of the decision making encounter
    print("""
You came into the city of Vizima and saw bounty notices posted asking for an experienced monster
hunter. As a witcher your livelihood is mostly slaying monsters in exchange for coins. You therefore, 
decided to track down this bounty.

         """)
    input(prompt ="<Press enter to continue>\n")
    starting_path()
    
####This is the 1st Scene    
def starting_path():
    print("""
    To start off, you need information about this quest. There might be places in this town where you can rest and gather 
    information. Looking through your vicinity right now, You spotted what seems to be the largest inn called the
    'Old Narakort Inn' and as usual at this hour it is full of people. Further down the street, a smaller Inn
    called "The Fox" was almost empty.
    
    Where will you go to rest and gather information?
    
    1) Proceed to Old Narakort Inn .
    2) Go further down the street and enter The Fox Inn
    3) Roam around the town to find more 
    """)
    
    
    choice = input(prompt ="Choose your action\n")
    choice = choice.lower()

    #####This choice is a loop 
    if "1" in choice or "proceed" in choice or "Old" in choice or "Narakort" in choice:
        print("""
         You walk towards the door and entered the inn. As soon as you enter, the noises from the people around
         dies down gradually as they started to stare at you with hostility. You walk down the aisle, towards 
         the bartender to ask him about any information but you were cut-off immediately as he beckons you aside. 
         He whispered his plea to you that you should leave. Judging by the reactions of his customers you are not
         wanted here and you might even start a trouble which is bad for his business. Out of respect you left 
         the building.
         """)
        input(prompt ="<Press enter to continue>\n")
        starting_path()
        
    ####This is the correct answer
    elif "2" in choice or "further" in choice or "fox" in choice or "down" in choice:
        print("""\n You went down the street and entered the Fox Inn.""")
        input(prompt ="<Press enter to continue>\n")
        fox()
        
    ####This is also a loop
    elif "3" in choice or "roam" in choice or "around" in choice or "find" in choice or "town" in choice:
        print("""
              You roamed around the town to look for other possible places and 
              you eventually found yourself standing the same spot as you were
              before. Nothing...
              """)
        input(prompt ="<Press enter to continue>\n")
        starting_path()
    
    ##### this is for debugging
    else:
        print("Invalid command. Please try again.")
        input(prompt ="<Press enter to continue>\n")
        starting_path()
    
##Second Scene   
def fox():
    print("""
    As you enter the tavern, you can feel all the stares and gazes on you. The people around you are giving off a 
    hostile vibe because of what you are. You maintained your composure as you walk towards the bartender.
    As you were about to talk to him, some aggressive rowdy thug cut you off and said," Oy you mutant son of a bitch!
    Getta out here before I stick your gut with my blade!". You turn around to face him. He is standing among what 
    seems to be his crew. What do you do?
    
    1) Avoid starting a fight at all cost and walk away
    2) Defend yourself and start a fight them
    3) Hypnotize them with the Axii sign spell
    """)
    ##### this is to have a consistent character sizes
    trouble = input(prompt = "> Choose your action \n")
    trouble = trouble.lower()
    
    ###This is the answer
    if "2" in trouble or "fight" in trouble or "defend" in trouble:
        print("""
        Your instinct tells you to defend yourself and as soon as you went into your fighting stance,
        the thug tries to sneak an attack with a dagger. With your insanely superhuman strength and speed, 
        you were able to react just in time and defend your self with a sword. When the chaos broke out it 
        can only be described as a one-sided bloodbath as the thug's crew engaged at you one by one. After 
        what seems to be just a few seconds of time passed, you stood among the bloody remains of what was 
        once were humans but now just a gruesome display of body parts. In just a few seconds, the guards 
        suddenly comes bursting in. After seeing the scene, they immediately knew what happened and felt
        fear down their spine. With their shaky voice they ordered you to stand down. You told them that 
        you will stand down as long as they let you follow them with no chains. The frightened guards had 
        no choice but to agree after seeing the bloody scene.
            """)
        input(prompt ="<Press enter to continue>\n")
        quest()
        
    #####This choice leads to mission failure
    elif "1" in trouble or "avoid" in trouble or "dismiss" in trouble:
        print("""
        \n As soon as you turn your back to leave, the thug immediately stabs you on the back. 
           Blood flows out from your body and you lost consciousness" 
         """)
        input(prompt ="<Press enter to continue>\n")
        fail()
        
    #####This choice will also lead to mission failure
    elif "3" in trouble or "Hypnotize" in trouble or "Axii" in trouble:
        print("""
              \n 
              You attempted to make a handsign in the air but the thug sneakily brought out a dagger and immediately
              sliced of your hand. Eventually, you lost consciousness due to blood loss.
              """)
        fail()
    
    
    ###This is for debugging that loops to the start
    else:
        print("Invalid command. Please try again and make sure to limit your input with the choices above.")
        input(prompt ="<Press enter to continue>\n")
        fox()
    
    
    
#3rd Scene         
def quest():
    print("""
    You were arrested and escorted to Velerad, the elderly castellan of Vizima. When asked to explain yourself
    you recounted your story and referred to the bounty notices you saw posted. Velerad then immediately knew 
    what is going, and tells you about the story related to the bounty. 
    """)
    
    input(prompt ="<Press enter to continue>\n")
    
    print("""
    In summary of what Velerand told you:
    Near the beginning of King Foltest's reign, he got his sister, Adda, pregnant. She and the baby died giving birth,
    both being buried in a double coffin. Seven years later, the daughter awoke and slaughtered the inhabitants of the
    palace. She had become a striga, some kind of a vicious monster. The monster wakes up at fullmoon and attack nearby
    settlements for its sustenance. Some of the nobles wants it to stay the same as this is a perfect display of of the 
    King's cowardice and weakness. In order to further investigate the matter you will have to follow certain sequence 
    of events to solve the matter.
    
    """)
    
    input(prompt ="<Press enter to continue>\n")
    
    print("""
    Arrange the following events in order to find a way to undo the curse of the princess. 
    Type the exact words among the choices for it to be considered valid answer. Extra spaces also affect your answer.
    (Hint: Read Velerand's story well based on its order)
    
    --> Seek the king and inquire about the curse and the striga
    --> Spend the night and investigate the old palace where the striga is sleeping
    --> Refuse the bribe offered by Lord Ostrit to spare the monster
    --> Go to see the miller regarding the attacks of the striga
    
    """)
    
    
    input(prompt ="<Press enter to start>\n")    
    
######The sequence of events for investigating the matter that the player must follow
    events=['seek the king and inquire about the curse and the striga',
            'spend the night and investigate the old palace where the striga is sleeping',
            'go to see the miller regarding the attacks of the striga',
            'refuse the bribe offered by lord ostrit to spare the monster']

    #####Every inputs will be put inside this empty list based on the order of inputs   
    sequence=[]
    
    
    print(" Enter the first sequence: \n")
    seq_1=str(input(prompt = " "))
    seq_1=seq_1.lower()
    print("\n Enter the second sequence: \n")
    seq_2=str(input(prompt = " "))
    seq_2=seq_2.lower()
    print("\n Enter the third sequence: \n")
    seq_3=input(prompt = " ")
    seq_3=seq_3.lower()
    print("\n Enter the fourth sequence: \n")
    seq_4=input(prompt = " \n")
    seq_4=seq_4.lower()
    
    #filling out the sequence
    sequence.append(seq_1)
    sequence.append(seq_2)
    sequence.append(seq_3)
    sequence.append(seq_4)
    
    
    #Confirm if final answer
    print("is this your final answer?\n")
    answer= input(prompt ="Yes or No? \n")
    answer=answer.lower()
    
    #######The correct answer(sequence) must be equal to the event list
    if answer == 'yes':
        if sequence == events:
            print("You were able to successfully investigate the quest")
            input(prompt ="<press enter to continue> \n")
        else:       
    #####This is to let players know where they got wrong
            print('You got lost along the way and failed to complete your task!')
            print("Correct answer: \n")
            print(events)
            print("Your answer: \n")
            print(sequence)
            fail()
            
    elif answer == 'no':
            print('Okay back to start!')
            quest()
    else:
            print("Invalid answer. Please type yes or no only")
            quest()

####Story continues after you suceeded
    print("""
    With your investigation you realized that that the only way to reverse the curse of the princess is to not allow
    the striga to sleep in her original coffin until the third crow of the rooster. You also found out that it was 
    Lord Ostrit, a previous suitor of Lady Adda, was the one who cast the curse upon the mother and child out of jelousy.
    """)
    
    input(prompt ="\n <Press enter to continue>\n")
    
    print("""
    With all this knowledge, you are now prepared to face the monster in the old palace. The king pleaded you to avoid
    killing her as much as possible if what you said about her being curable is true. After preparing your necessary
    equipments, you drank some sort of a mutagenic concoctions to augment your fighting skills. You then set out to 
    reverse the princess' curse in the palace.
    """)
    
    input(prompt ="\n <Press enter to continue>\n")
    old_palace()
    
    
###4th Scene
def old_palace():
    print("""

    As you finished your preparations, you ventured towards the castle gates and went straight to the 
    bedroom of the late princess Adda. There you meditated and waited until the full moon rises...
          
          """)
    
    input(prompt ="\n <Press enter to continue>\n")
    
    print("""
     As soon as the moon rays shone through the windows, you heard a rumble beneath the palace. At that exact moment
     you knew the monster has awakened from its slumber. You readied yourself as you heard the aggressive footsteps
     of a quadrupedal creature coming closer.Then you saw the striga walking like a hunter stalking its prey. 'A foul,
     well-muscled monster, the cursed princess runs about on all fours. Her jaw extends from ear to ear, and her fingers 
     end in claws capable of tearing a man to shreds. Her breasts and red hair are the sole reminders that the striga is, 
     in fact, a young girl possessed by an evil curse.'
     
     """)
    
    input(prompt ="\n <Press enter to continue>\n")
          
    print("""  
    As you recall, the only way to save it is to avoid killing it. Whatever the monster is going to do, you must
    ready to counter every possible attack it throws on you until you weaken it. You readied yourelf for battle as
    the striga engages you.
         """)
    
    
    ####### These are the monster's moveset
    
    monster_behaviour = ['The striga tensed and will likely try to leap!',
                         'The striga raised its arms to swipe you with its mighty claws!', 
                         'The striga opens its mouth in an attempt to bite you!', 
                         'The striga runs backward to create distance and then charges towards you in attempt to knock you off!', 
                         'The striga tries to intimidate you with its terrifying roar!']
    ####### Player must match the proper fighting move given in the choices in order to counter monster's attacks
    ####### Player will fight the striga 4 times before it is weakened and one wrong choice can spell the end for the player
    
    engages = 4
    
    while engages >0:
        engage=random.choice(monster_behaviour)
        input(prompt ="<Press enter to engage the striga>\n")
        print(engage)
        print(" ")
        
        if engage == monster_behaviour[0]:
            print("""
            You need to immobilize him... what should you do?
            
            1) Parry and attack with your silver sword
            2) Use silver chain
            3) Punch it using your silver knuckles
            4) Use Aard Sign or Force Magic to push away
            5) Use a charm magic to project its hate and malice back at it
            """)
            immobilize = input(prompt ="> \n")
            immobilize = immobilize.lower()
        
        ################If the monster tries to leap use chain to hold it down
        
            if '2' in immobilize or 'chain' in immobilize:
                print("""
                The chained whistled through the air, coiling like a snake and twined itself around the monster.
                The striga tumbles and is now stuck on the ground.
                """)
                engages -= 1
                
            else:
                print("The striga was able to read your incoming move!")
                print("""
                It dodges whatever you are trying to do and lands on your body killing you with one swipe 
                of its claw.
                """)
                dead()
                
      #####################If monster uses its claws counter it with your sword          
        elif engage == monster_behaviour[1]:
            print("""
            You have to counter it in someway! What do you do?
            
            1) Parry and attack with your silver sword
            2) Use silver chain
            3) Punch it using your silver knuckles
            4) Use Aard Sign or Force Magic to push away
            5) Use a charm magic to project its hate and malice back at it
            """)
            parry= input(prompt ="> \n")
            parry= parry.lower()
            
            if '1' in parry or 'parry' in parry or 'silver sword' in parry:
                print("""
                Fortunately, because of your superhuman speed and strength, you were able to parry 
                its attack with your silver sword with enough strength in just the nick of time. 
                You then followed it up with a quick light slash which grazed her skin.
                """)
                engages -= 1
                
            else:
                print("Unfortunately, the striga was just too fast and you were not able block its attack!")
                print("""
                You took the full force of her attack, heavily wounding you as you flew
                towards the wall making you fall unconscious.
                """)
                dead()
                
    ######################## When it tries to bite you counter it with a punch
        elif engage == monster_behaviour[2]:
            print("""
            Your instinct is telling you to divert its head with your quickest possible attack
            
            1) Parry and attack with your silver sword
            2) Use silver chain
            3) Punch it using your silver knuckles
            4) Use Aard Sign or Force Magic to push away
            5) Use a charm magic to project its hate and malice back at it
            """)
            punch= input(prompt ="> \n")
            punch= punch.lower()
            
            if '3' in punch or 'punch' in punch or 'knuckle' in punch:
                
                print("""
                You punched the side of its head as soon as it tries to close the gap between your face and its jaws.
                Your punch was strong enough to divert its head towards a wall crashing against it. 
                The monster stood up and regained its composure
                """)
                engages -= 1
                
            else:
                print("""
                It was too fast and close enough that you were not able to react immediately.
                You never saw the light of the day as its jaws came towards your face.
                """)
                dead()
                
    ###########When monster charges at you counter it with Aard magic(force)!
        elif engage == monster_behaviour[3]:
            print("""
            Your instinct is telling you that you cannot handle its full force, you somehow need to stop its charge!
            
            1) Parry and attack with your silver sword
            2) Use silver chain
            3) Punch it using your silver knuckles
            4) Use Aard Sign or Force Magic to push away
            5) Use a charm magic to project its hate and malice back at it
            """)
            aard= input(prompt ="> \n")
            aard= aard.lower()
            
            if '4' in aard or 'aard' in aard or 'force magic' in aard or 'away' in aard:
                print("""
                As soon as you saw it charging, you immediately cast the Aard Sign. 
                A sudden burst of force magic made the monster fly away and crash towards a wall.
                The monster shakes its body to remove debris and then stands up to regain its composure
                """)
                engages-=1
            else:
                print("""
                Whatever attempt you did was ineffective as the monster closed the gap instantly.
                Taking the full force of the charge, you were sent flying towards a wall making you unconscious.
                """)
                dead()
                
    ##########Monster is trying to break your will but counter it with charm magic        
        elif engage == monster_behaviour[4]:
            print("""
            Whatever it is trying to do seems to affect your will to fight.
            What do you do?
            
            1) Parry and attack with your silver sword
            2) Use silver chain
            3) Punch it using your silver knuckles
            4) Use Aard Sign or Force Magic to push away
            5) Use a charm magic to project its hate and malice back at it
            """)
            sorcery= input(prompt ="> \n")
            sorcery= sorcery.lower()
            
            if '5' in sorcery or 'charm' in sorcery or 'project' in sorcery or 'hate and malice' in sorcery:
                print("You broke down the monster's mental will making it flinch in pain and terrified for a few seconds.")
                engages -= 1
                
            else:
                print("""
                Your mental will broke down and you decided to flee the fight but the monster was fast to
                knock you off and drag you away towards its nest.
                """)
                dead()
                
                break
    
    #####Once weakened the concluding story continues and the game ends 
    
    input(prompt ="\n <Press enter to continue> \n")
    
    print("""
    You finally managed to weaken the striga and it is now attempting to flee away as soon it realizes that 
    the dawn is breaking. It is running towards the coffin to get back to its slumber before the sun comes out
    but due to its injuries you were able to over take it. You managed to push away the monster before it gets 
    inside the coffin. You went inside the coffin to fortify and shut the lid with magic to prevent the monster
    from coming in. You can hear the monster thrashing at the coffin lid but it has no effect. You then waited 
    until hear "the third crow of the rooster". Then, everything went silent...
    """)
    
    
    input(prompt ="\n <Press enter to continue> \n")
    
    print("""
    You come out of the coffin and you saw a naked girl lying on one side covered in mud and dirt. You realized your mission
    was a success and was able to revert the striga back to the princess. You went closer to bring her back to the king
    but as soon as you touch her, the girl turns around and lunges herself at you. She bit a flesh on the side of your
    neck. The last image you saw was what seems to be a teenage girl with a mind of a 3 year old child with an animalistic
    behaviour cowering in fear. You fell unconscious as your blood continous to flow out...
    """)
    
    input(prompt= "\n <Press enter to continue> \n")
    
    
    success()
           


        
####Mission success
def success():
    print("""
Congratulations you were able to save the princess! 

Eventually the king and his men rescued you and the princess just in time
before you die of blood loss.

Your reward will be given to you as soon as you wake up!
          """)
    input(prompt ="<Press enter to continue>\n")
    print(""" 
          Would you like to play again?
          
          Press 1 to play again
          Press 2 to end the game
          """)
    
    
    play = input(prompt = "> ")
    
    if play == '1':
        start() 
        # returns at start of the game

    elif play == '2':
        print("""Thanks for playing!""")
        exit(0)
        # the game ends
    
    else:
        print("Invalid entry. Please try again.\n")
        input(prompt ='<Press any key to continue>\n')
    exit(0)

####Specifically made for the 4th scene
def dead():
    print("""\nYou have failed to save the princess. 
          Would you like to try again?
          
          Press 1 to play again
          Press 2 to end the game
          """)
    
    
    play = input(prompt ="> ")
    
    if play == '1':
        start() 
        # returns at start of the game

    elif play == '2':
        print("""Thanks for playing!""")
        exit(0)
        # the game ends
    
    else:
        print("Invalid entry. Please try again.\n")
        input(prompt ='<Press any key to continue>\n')
    
    
    
def fail():
        print("""\nYou have failed to complete your Journey. 
          Would you like to try again?
          
          Press 1 to play again
          Press 2 to end the game
          """)
    
    
        play = input(prompt = "> ")
    
        if play == '1':
                  start() 
        # goes back to start of the game

        elif play == '2':
            print("""Thanks for playing!""")
            exit(0)# the game ends
    
        else:
            print("Invalid entry. Please try again.\n")
            input(prompt ='<Press any key to continue>\n')

start()
```
